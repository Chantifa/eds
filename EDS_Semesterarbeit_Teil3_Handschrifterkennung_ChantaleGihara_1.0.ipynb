#### Was ist ein neuronales Netz eigentlich genau?
Ein neuronales Netz besteht aus Perceptronen. Wie funktionieren Perceptronen? Ein Perceptron nimmt mehrere binäre Eingänge, $x_1$, $x_2$, ..., und produziert einen einzigen binären Ausgang:

 [8]

In dem gezeigten Beispiel hat das Perceptron drei Eingänge, $x_1$ , $x_2$, $x_3$. Im Allgemeinen könnte es mehr oder weniger Eingänge haben. Dazu werden Gewichtungen für den Input vergeben, $w_1$, $w_2$,..., reelle Zahlen, die die Bedeutung der jeweiligen Eingänge für den Ausgang ausdrücken. Die Ausgabe des Neurons, 0 oder 1, wird dadurch bestimmt, ob die gewichtete Summe $\sum\limits_{j}$ $w_j$ * $x_j$ kleiner oder grösser als irgendein Schwellenwert ist. Genau wie die Gewichte ist der Schwellenwert eine reelle Zahl, die ein Parameter des Neurons ist. Um es in präziseren algebraischen Begriffen auszudrücken:

Das ist das grundlegende mathematische Modell bezüglich der Funktionsweise eines Perceptrons . Man kann über das Perceptron denken, dass es ein Gerät ist, das Entscheidungen trifft, indem es Beweise abwägt. [7]

Natürlich ist das Perceptron kein vollständiges Modell der menschlichen Entscheidungsfindung, aber es kann dabei unterstützen subtile Entscheidungen zu veranschaulichen, wie folgendes Beispiel aufzeigt::



In diesem Netzwerk sieht man die erste Säule der Perceptrons, bezeichnet als «erste Schicht». Hier sind drei sehr einfache Entscheidungen zu treffen, indem sie die eingegebenen Beweise abwägt.

***Was ist mit den Perceptrons in der zweiten Schicht?***

Jedes dieser Perceptrons trifft eine Entscheidung, indem es die Ergebnisse der ersten Schicht der Entscheidungsfindung abwägt. Auf diese Weise kann ein Perceptron in der zweiten Schicht eine Entscheidung auf einer komplexeren und abstrakteren Ebene treffen als die Perceptrons in der ersten Schicht. Und noch komplexere Entscheidungen kann das Perceptron in der dritten Schicht treffen. Auf diese Weise kann ein vielschichtiges Netzwerk von Perceptrons eine anspruchsvolle Entscheidungsfindung durchführen. Es gibt, wie bereits oben beschrieben, nur einen Ausgang, die Pfeile für mehrere Ausgänge dienen lediglich dazu anzuzeigen, dass der Ausgang eines Perceptrons als Eingang für mehrere andere Perceptrons verwendet wird.

Zunächst muss eine Funktion definiert werden, welche die Anfrage an Microsoft Azure sendet und das in der Antwort erhaltene Ergebnis zurückgibt. Hier wird die Methode requests.request verwendet, um eine Verbindung mit dem Azure-Server mit den folgenden 6 Parametern herzustellen.

* Methode: Methode für das neue Request-Objekt. In diesem Fall ist es 'post'.

* url: URL für das neue Request-Objekt. Dies wurde oben bereits als 'url' definiert.

* params: Boolesches Objekt für Handschrift oder gedruckten Text.

* Daten: Bilddaten-Objekt

* Json: Ein serialisierbares JSON-Python-Objekt, das im Hauptteil der Anfrage gesendet werden soll.

* Header: Abonnementschlüssel und die Datentyp-Anforderungen

Wenn der Server die Anfrage angenommen hat, gibt er den Statuscode zurück und fügt einen "location"-Header hinzu. Location gibt die Adresse an, von der wir unser Ergebnis erhalten können.

Wichtiger Hinweis: 'json' und 'data' sind beide Eingabe-Bilddatenobjekte. Der Unterschied besteht darin, dass das 'data'-Objekt für Bilder verwendet wird, die auf der Festplatte gespeichert sind, während das 'json'-Objekt für Bilder verwendet wird, die über eine URL abgerufen werden.

´
Parameter:
json: Wird verwendet, wenn Bilder von seiner URL verarbeitet werden. Siehe API-Dokumentation
Daten: Wird bei der Verarbeitung von von der Festplatte gelesenen Bildern verwendet. 
Kopfzeilen: Wird zur Übergabe der Schlüsselinformationen und der Datentyp-Anforderung verwendet

[6]
´

```python
def proc_request( json, data, headers, params ):
    retries = 0
    result = None
    while True:
        response = requests.request( 'post', url, json = json, data = data, headers = headers, params = params )
        if response.status_code == 429:
            print( "Message: %s" % ( response.json() ) )
            if retries <= max_retries: 
                time.sleep(1) 
                retries += 1
                continue
            else: 
                print( 'Fehler: failed after retrying!' )
                break
        elif response.status_code == 202:
            result = response.headers['Operation-Location']
        else:
            print( "Error code: %d" % ( response.status_code ) )
            print( "Message: %s" % ( response.json() ) )
        break      
    return result
```

Hilfsfunktion zum Abrufen von Textergebnissen vom Einsatzort

Parameter:
1. location: location, um das Textergebnis zu erhalten, siehe API-Dokumentation
2. Kopfzeilen: Werden zur Weitergabe der Schlüsselinformationen verwendet

[6]

```python
def getTextResult( location, headers ):

    retries = 0
    result = None

    while True:
        response = requests.request('get', location, json=None, data=None, headers=headers, params=None)
        if response.status_code == 429:
            print("Message: %s" % (response.json()))
            if retries <= max_retries:
                time.sleep(1)
                retries += 1
                continue
            else:
                print('Error: failed after retrying!')
                break
        elif response.status_code == 200:
            result = response.json()
        else:
            print("Error code: %d" % (response.status_code))
            print("Message: %s" % (response.json()))
        break

    return result
```

Anzeige der erhaltenen Ergebnisse auf dem Eingabebild

1. Zuerst wird das Bild mit der Datei-Öffnen- und Lesemethode geladen.
2. Dann werden die Parameter für die REST-API festgelegt.
3. Danach wird die Anfrage an den Azure-Server gestellt.
4. Das JSON-Objekt vom Azure-Server wird geholt.
5. Das Originalbild wird geladen und das Ergebnis wird darauf angezeigt.

[5]

```python
def showImage( result, img ):
    img = img[:, :, (2, 1, 0)]
    fig, ax = plt.subplots(figsize=(12, 12))
    ax.imshow(img, aspect='equal')

    lines = result['recognitionResult']['lines']

    for i in range(len(lines)):
        words = lines[i]['words']
        for j in range(len(words)):
            tl = (words[j]['boundingBox'][0], words[j]['boundingBox'][1])
            tr = (words[j]['boundingBox'][2], words[j]['boundingBox'][3])
            br = (words[j]['boundingBox'][4], words[j]['boundingBox'][5])
            bl = (words[j]['boundingBox'][6], words[j]['boundingBox'][7])
            text = words[j]['text']
            x = [tl[0], tr[0], tr[0], br[0], br[0], bl[0], bl[0], tl[0]]
            y = [tl[1], tr[1], tr[1], br[1], br[1], bl[1], bl[1], tl[1]]
            line = Line2D(x, y, linewidth=3.5, color='red')
            ax.add_line(line)
            ax.text(tl[0], tl[1] - 2, '{:s}'.format(text),
            bbox=dict(facecolor='blue', alpha=0.5),
            fontsize=14, color='white')

    plt.axis('off')
    plt.tight_layout()
    plt.draw()
    plt.show()
```

Zunächst sind die Verbindungen zwischen den Neuronen im Netz zufällig, und das Netz ist nicht in der Lage, etwas Sinnvolles zu tun. Es liefert lediglich eine zufällige Antwort. 

Die Antwort wird mit dem Etikett verglichen. In diesem Fall unterscheidet sich die Antwort vom Label, und dem neuronalen Netz wird eine gewisse Rückmeldung gegeben, damit es sich verbessern kann. Die Verbindungen zwischen den Neuronen werden modifiziert, wobei diejenigen Neuronen bevorzugt werden, die dazu neigen, eine korrekte Antwort zu geben. 

Nach der Modifikation werden die nächsten Beispiele betrachtet, und das neuronale Netz lernt in einem iterativen Prozess. 

Die Anzahl der Trainingsbeispiele, die für ein richtiges Training des Netzes erforderlich sind, liegt in der Grössenordnung von einigen Hundert für Netze mit einer einfachen Architektur bis zu Millionen für komplexe Netze. [11]

Die Eingabeschicht erfordert ein 1-dimensionales Array bei der Eingabe, aber die vorgegebenen Bilder sind 2D. Die Bilder müssen somit abgeflacht werden:

```python
y = digits.target
x = digits.images.reshape((len(digits.images), -1))
x.shape
```

Nun sind es 1797 verflachte Bilder. Die zwei Dimensionen der 8x8-Bilder wurden zu einer einzigen Dimension zusammengelegt, indem die Reihen von 8 Pixeln nacheinander geschrieben wurden. Das erste Bild, das zuvor betrachtet wurde, wird jetzt durch ein 1-D-Array mit 8x8 = 64 Slots dargestellt.

Die Daten werden in eine Trainings- und eine Teststichprobe aufgeteilt:

Die ersten 1000 Bilder und Etiketten werden für die Tests verwendet. Der Rest des Datensatzes wird später verwendet, um die Leistung des Netzwerks zu testen.

```python
x_train = x[:1000]
y_train = y[:1000]
x_test = x[1000:]
y_test = y[1000:]
```

Jetzt kann das neuronale Netz erstellt werden. Hier wird eine Schicht mit 15 Neuronen verwendet, und Scikit -Learning[10] weiss wie herauszufinden, wie viele Zahlen in der Eingabe- und Ausgabeschicht zu verwenden sind.

```python
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(hidden_layer_sizes=(15,), activation='logistic', alpha=1e-4,
                    solver='sgd', tol=1e-1, random_state=1, max_iter=500,
                    learning_rate_init=.1, verbose=True)

mlp.fit(x_train,y_train)
```

Da das neuronale Netz einfach ist und der Eingabedatensatz klein ist, war der Lernprozess schnell. Nun, da das Netzwerk trainiert wurde, ist erkennbar, was es über unsere Testbilder aussagen kann. Dabei werden die ersten 50 Beispiele in den Tests betrachtet. Diese Vorhersagen sollten den Zielen der Lernstichprobe ziemlich nahekommen, hierzu werden nun die Arrays verglichen.

Um noch die verschiedenen Layers (hidden_layers) und das gelernet etwas besser zu verstehen, möchte ich hier noch etwas mehr auf das Tests und das Training eingehen. 
Es gibt drei Arten von Solver: "lbfgs", "sgd" und "adam", wobei "adam"[10]. Für eher kleinere Datensätze eignet sich "lbfgs", weil es schneller konvergiert. Nach der Festlegung der Klassifikation wird das Modell an die Datenmatrix X und die Ziele y angepasst. In diesem Beispiel werden die verarbeiteten Bilder von MNIST als Array mit X und das Zielbild mit y deklariert. Die Methode train_test_split() mischt die Daten zufällig. Es ist wichtig, dass ein Teil der Daten zum Testen verwendet wird, da auf diese Weise das Risiko einer Überanpassung vermieden werden kann. Die grosse Herausforderung bei der Erkennung einer Überanpassung besteht darin, dass es schwierig ist, herauszufinden, wie gut ein Modell ist, ohne es vorher getestet zu haben.

```python
from sklearn.model_selection import train_test_split
digits = datasets.load_digits()
X, y = digits.data / 255., digits.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25,random_state=42)
mlp = MLPClassifier(hidden_layer_sizes=(100,), max_iter=100, alpha=1e4,
solver='lbfgs', verbose=10, tol=1e4, random_state=1,learning_rate_init=.1)
mlp.fit(X_train, y_train)
print()
print("Training score: {0}".format(mlp.score(X_train, y_train)))
print("Test score: {0}".format(mlp.score(X_test, y_test)))
```

```python
digits = datasets.load_digits()
X, y = digits.data / 255., digits.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25,random_state=42)
mlp = MLPClassifier(hidden_layer_sizes=(500,), max_iter=500, solver='lbfgs', random_state=1, learning_rate_init=.1)
mlp.fit(X_train, y_train)
print()
print("Training set score: {0}".format(mlp.score(X_train, y_train)))
print("Test set score: {0}".format(mlp.score(X_test, y_test)))
predicted = mlp.predict(X_test)
for i in range(1):
    image = np.reshape(X_test[i], (8, 8))
    plt.figure(figsize=(2, 2))
    plt.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    plt.show()
print('Ground Truth: {0}'.format(y_test[i]))
print('Predicted: {0}'.format(predicted[i]))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predicted))
```

Wie zu sehen war sind die OCR Technologien lernfähig, weshalb API's verwendet werden sollten, die schon viele Erfahrungen haben und somit eine grosse Anzahl Perceptronen ausgewertet haben. Es muss dem Answender jedoch klar  sein, wie die APIs funktionieren, dafür eignet sich scikit-learn[10]. Eine bestehenden API, wie hier mit Microsoft Azure zu verwenden ist vorteilhaft, da diese eine der meistverwendeten ist und somit schon über eine sehr genaue Trefferquote verfügt.
